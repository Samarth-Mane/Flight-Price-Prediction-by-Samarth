```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
train_data = pd.read_excel(r"Data_Train1.xlsx")
pd.set_option('display.max_columns',None) #display all the columns from the excel.
train_data.head()
```

```python
train_data.info()   #data-type - object which means the String such as flight name and other informations
```

```python
train_data["Duration"].value_counts() #it tell that like 2h50m duration are 550 flight and 1h30m duration have 386 flights and so on
```

```python
train_data.dropna(inplace = True)  #it will tell the total numbers of null values and there are no null values in 10682 elements and 11 columns
```

```python
train_data.isnull().sum()
```

```python
train_data["Journey_day"] = pd.to_datetime(train_data["Data_of_journey"], format="%d/%m/%Y").dt.day
#the model will not be able to understand the string date&time formate for that by (to_datetime) we get the day by dt,day
```

```python
train_data["Journey_month"] = pd.to_datetime(train_data["Data_of_journey"], format = "%d/%m/%Y").dt.month
#the model will not be able to understand the string date&time formate for that by (to_datetime) we get the month by dt,month
```

```python
train_data.drop(["Data_of_journey"], axis = 1, inplace = True)
# Since we have converted Date_of_Journey column into integers journey_day,journey_month, Now we can drop it.
```

```python
print(train_data.shape)
print(train_data.columns)
```

```python
train_data = train_data.loc[:, ~train_data.columns.str.contains('^Unnamed')]
```

```python
train_data.head()
```

```python
train_data.head(5)
```

```python
#(DEPARTURE TIME)
# Departure time is when a plane leaves the gate. 
# Similar to Date_of_Journey we can extract values from Dep_Time

# Extracting Hours
train_data["Dep_hour"] = pd.to_datetime(train_data["Dep_Time"]).dt.hour

# Extracting Minutes
train_data["Dep_min"] = pd.to_datetime(train_data["Dep_Time"]).dt.minute

# Now we can drop Dep_Time as it is of no use
train_data.drop(["Dep_Time"], axis = 1, inplace = True)

```

```python
train_data.info()
```

```python
# Print column names
print(train_data.columns)

# Check for missing or empty values
print(train_data.isnull().sum())

# Look at the first few rows
print(train_data.head(10))
```

```python
train_data.head()
```

```python
#(ARRIVAL TIME)
# Arrival time is when the plane pulls up to the gate.
# Similar to Date_of_Journey we can extract values from Arrival_Time

# Extracting Hours
train_data["Arrival_hour"] = pd.to_datetime(train_data.Arrival_Time).dt.hour #convert the arrival time into the date time here it  converts into hours

# Extracting Minutes
train_data["Arrival_min"] = pd.to_datetime(train_data.Arrival_Time).dt.minute

# Now we can drop Arrival_Time as it is of no use
train_data.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
# Time taken by plane to reach destination is called Duration
# It is the differnce betwwen Departure Time and Arrival time


# Assigning and converting Duration column into list
duration = list(train_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
# Adding duration_hours and duration_mins list to train_data dataframe

train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_mins
```

```python
train_data.drop(["Duration"], axis = 1, inplace = True)
```
